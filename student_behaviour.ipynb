```python
df.drop('Commerce', axis=1, inplace=True)
df.drop('Video Games', axis=1, inplace=True)
df.drop('Night', axis=1, inplace=True)
```

```python
for i in sorted(df['Department'].unique()):
    df.loc[df['Department'] == i, i] = 1
```

```python
for i in sorted(df['hobbies'].unique()):
    df.loc[df['hobbies'] == i, i] = 1
```

```python
for i in sorted(df['prefer to study in'].unique()):
    df.loc[df['prefer to study in'] == i, i] = 1
```

```python
df.drop('Department', axis=1, inplace=True)
df.drop('hobbies', axis=1, inplace=True)
df.drop('prefer to study in', axis=1, inplace=True)
```

```python
df = df.fillna(0)
```

```python
df.loc[df['Travelling Time '] == '0 - 30 minutes', 'Travelling Time '] = '1'
df.loc[df['Travelling Time '] == '30 - 60 minutes', 'Travelling Time '] = '2'
df.loc[df['Travelling Time '] == '1 - 1.30 hour', 'Travelling Time '] = '3'
df.loc[df['Travelling Time '] == '1.30 - 2 hour', 'Travelling Time '] = '4'
df.loc[df['Travelling Time '] == '2 - 2.30 hour', 'Travelling Time '] = '5'
df.loc[df['Travelling Time '] == '2.30 - 3 hour', 'Travelling Time '] = '6'
df.loc[df['Travelling Time '] == 'more than 3 hour', 'Travelling Time '] = '7'
```
